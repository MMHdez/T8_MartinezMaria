```python
# importar librerías para análisis de datos y manipulación de fechas
import pandas as pd
import numpy as np
from datetime import date
import numpy_financial as npf
```

```python
# establecer los datos para la cuenta de ahorro
tasa_interes = 0.04
years = 5
Abono_year = 12
inicial = 10000
meta = 85000
fecha_inicio = (date(2019,1,1))
```

```python
# Realizar el cálculo de abonos mensuales iguales para llegar a la meta de ahorro en el tiempo deseado
Abono = -npf.pmt(tasa_interes/Abono_year,years*Abono_year,(meta-inicial))
Abono
```

```python
# calcular el periodo
periodo = 50

# calcular el interés
ipmt = npf.ipmt(tasa_interes/Abono_year, periodo, years*Abono_year, (meta-inicial))

#calcular el inicial
ppmt = npf.ppmt(tasa_interes/Abono_year, periodo, years*Abono_year, (meta-inicial))

print (ipmt, ppmt)
```

```python
# Estimar el rango de fechas en la tabla
rng = pd.date_range(fecha_inicio, periods= years * Abono_year, freq='M')
rng.name = 'Fecha del abono'

# Establecer columnas de abonos y extracto de cuenta en las fechas respectivas
df = pd.DataFrame(index = rng, columns = ['Abono','Balance'],dtype='float')

# Colocar el índice con los periodos y no con las fechas
df.reset_index(inplace=True)

df.index+=1
df.index.name='Periodo'
df
```

```python
npf.ppmt(tasa_interes/Abono_year, 1, years * Abono_year, Abono)
```

```python
# Los abonos son iguales
df['Abono'] = Abono

# Vector indizador de los periodos de los abonos
index_vector = np.arange(1,len(rng)+1)

# Extracto de la cuenta mes a mes
df["Balance"] = inicial * (1 + tasa_interes/Abono_year)**index_vector + \
                Abono * (((1 + tasa_interes/Abono_year)**index_vector - 1)/(tasa_interes/Abono_year))

df
```
